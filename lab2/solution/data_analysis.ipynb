### Eksploracyjna analiza danych

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

data_train = pd.read_csv("df_train.csv")
y_train = data_train["MedHouseVal"]
X_train = data_train.drop(columns=["MedHouseVal"])
# X_test = pd.read_csv("X_test.csv")
X_train.head()
```

##### Macierz korelacji

```python
%matplotlib inline
corr_matrix = data_train.corr()
heatmap = sns.heatmap(corr_matrix, cmap="crest", annot=True, fmt=".2f", linewidth=.5)
heatmap.set(title="Correlation between predictors")
plt.show()
```

##### Rozkłady parametrów

```python
duration_hist = sns.violinplot(data_train["MedInc"])
duration_hist.set(ylabel= "Median Income")
plt.title("Median income distribution")
plt.show()
```

```python
duration_hist = sns.violinplot(data_train["HouseAge"])
duration_hist.set(ylabel= "House Age Income")
plt.title("House age distribution")
plt.show()
```

```python
duration_hist = sns.violinplot(data_train["AveRooms"])
duration_hist.set(ylabel= "Average room amount")
plt.title("Avg rooms distribution")
plt.show()
```
