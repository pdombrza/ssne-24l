### Data analysis

```python
from matplotlib import pyplot as plt
import seaborn as sns
from recursion import load_data
from collections import Counter
```

```python
data = load_data('train.pkl')

classes = [label for _, label in data]
class_count = dict(Counter(classes))
class_count
```

```python
composers = {"Bach": 0, "Beethoven": 1, "Debussy": 2, "Scarlatti": 3, "Victoria": 4}
composers_count = {}
for key, val in composers.items():
    composers_count[key] = class_count[val]
print(composers_count)

fig, ax = plt.subplots()
ax.bar(composers_count.keys(), composers_count.values())
plt.show()
```
